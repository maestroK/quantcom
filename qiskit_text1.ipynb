```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram
#creating a circuit with eight qubits
qc = QuantumCircuit(8)
#measuring all qubits
qc.measure_all()
#visualizing the circuit
qc.draw()
```

```python
#giving the circuit an initial state. Qubits are always initialized to give the output 0.
qc.draw(initial_state = True)
```

```python
#since we didnt do anything to the qubit, the final result must be the same as initial state ie 0. we can see that by plotting a histogram
# thoughh you can plot histogram for vectors using statistics = qc.sample_counts(100). for circuits, you need the aer function
#The reason for running many times and showing the result as a histogram is because quantum computers may have some randomness in their results
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
#we must add the x gate before measuring otherwise the x applies after measurement ie has no effect
qc1 = QuantumCircuit(8)
qc1.x(7)
qc1.draw()
```

```python
#before we plot the histogram, we must measure the qubits
#writing 7 in qubits
qc2 = QuantumCircuit(8)
#you could do qc2.x() qc2.x() qc2.x() but instead try. remembering that last qubit is the first bit
qc2.x([5, 4, 2, 1, 0])
qc2.draw()
```

```python
#now remembering if you want to plot the histogram, first you need to measure them
qc2.measure_all()
qc2.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc2).result()
counts = result.get_counts()
plot_histogram(counts)
```
